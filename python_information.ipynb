A string is a sequence of characters enclosed in single or double quotes.

Boolean data type represents True or False values.

#9. Python Type Conversion
###What is Python Type Conversion?
Type conversion is the process of changing one data type into another data type.  
In simple words, type conversion means converting the value of one data type into another data type.  
Example

    x = 10
    y = "20"
    z = x + int(y)
    print(z)
Output:
30

Explanation:  
x is an integer  
y is a string  
int(y) converts the string "20" into integer 20  
Then Python adds 10 + 20

###Types of Type Conversion in Python:
####1. Implicit Type Conversion

Implicit conversion is automatically done by Python when converting a smaller data type into a larger data type.

Example:

    x = 10
    y = 5.5
    result = x + y
    print(result)
Output:
15.5  

Explanation:  
x is int  
y is float  
Python automatically converts int → float  

####2. Explicit Type Conversion (Type Casting)

Explicit conversion is manually done by the programmer using functions like int(), float(), str().

Example:

    x = "50"
    y = int(x)
    print(y)
Output:
50

Explanation:  
"50" is a string  
int(x) converts it into integer

####Common Type Conversion Functions

int() : Converts value to integer  
float() :	Converts value to float  
str() :	Converts value to string  
bool() :	Converts value to boolean

#12. Python Namespace and Scope
###1. Python Namespace

A namespace is a container that stores the names of variables, functions, and objects in a program.  
It helps Python organize names and avoid conflicts between variables with the same name.  
In simple words, a namespace is a collection of names and their corresponding objects.

Example:

      x = 10
      def show():
      y = 20
      print(y)
    print(x)
Explanation:


x belongs to the global namespace  
y belongs to the local namespace inside the function

###Types of Namespace:
####1. Built-in Namespace
Contains built-in functions and keywords provided by Python.  
Example:

    print()
    len()
    type()
####2. Global Namespace
Contains variables defined outside functions in a program.  
Example:

    x = 10
    def show():
      print(x)

Here x belongs to the global namespace.

####3. Local Namespace
Contains variables defined inside a function.

Example:

    def show():
      y = 20
      print(y)

Here y belongs to the local namespace.

###2. Python Scope

Scope refers to the area of the program where a variable can be accessed.  
In simple words, scope defines where a variable is available in the program.

####Types of Scope:
####1. Local Scope

Variables defined inside a function can only be used inside that function.

Example:

    def show():
      x = 5
      print(x)
      show()
####2. Global Scope

Variables defined outside functions can be used throughout the program.

Example:

    x = 10
    def show():
      print(x)
    show()

#13.Interpreter vs Compiler
###Interpreter
An interpreter is a program that executes code line by line and converts it into machine language during runtime.  
It reads one statement at a time and immediately executes it.
###Example  
Python uses an interpreter.
###Features of Interpreter:

Executes code line by line  
Stops when an error occurs  
Easier to debug  
Slower execution compared to a compiler

Example:

    print("Hello")
    print("World")

The interpreter runs each line one by one.

###Compiler

A compiler is a program that translates the entire source code into machine code at once before execution.  
After compiling, the program can run without the compiler.

###Example:
Languages like C and C++ use a compiler.

###Features of Compiler:  
Translates the entire program at once  
Errors are shown after compilation  
Execution is faster  
Requires compilation before running

| Interpreter                | Compiler                        |
| -------------------------- | ------------------------------- |
| Executes code line by line | Converts entire program at once   |
| Stops when an error occurs | Shows errors after compilation    |
| Slower execution           | Faster execution                |
| Example: Python            | Example: C, C++                 |
