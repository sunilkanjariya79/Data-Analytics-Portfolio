# Insight 1:

```python
# plotting a bar chart for day and its count
plt.figure(figsize=(8,5))
sns.countplot(x='day', data=df, palette='pastel', hue='day', legend=False)

for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Day-wise Customer Count",size=12)
```

```python
# plotting a bar chart for time and its count
plt.figure(figsize=(6,4))
sns.countplot(x='time', data=df, palette='Set3', hue='time', legend=False)

for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Time Distribution",size=12)
```

# Insight 2:
From above graphs we can see that most customers visit during weekends (Sat & Sun).
Also, dinner time has more customers compared to lunch.

```python
# plotting distribution of total bill
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='total_bill', bins=20, kde=True, hue='sex', palette='Set2')

plt.title("Total Bill Distribution")
```

```python
# plotting distribution of tip
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='tip', bins=20, kde=True, hue='smoker', palette='coolwarm')

plt.title("Tip Distribution")
```

# Insight 3:
 From above graphs we can see that most total bills are in mid-range.
 Tips are generally smaller compared to total bill,
but follow a similar distribution pattern.

```python
# plotting total bill vs tip
plt.figure(figsize=(8,5))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex', palette='Set1')

plt.title("Total Bill vs Tip")
```

```python
# plotting size vs total bill
plt.figure(figsize=(8,5))
sns.scatterplot(x='size', y='total_bill', data=df, hue='smoker', palette='cool')

plt.title("Group Size vs Total Bill")
```

# Insight 4:
From above graphs we can see that higher total bills result in higher tips.
 Also, as group size increases, total bill also increases.

```python
# plotting average total bill by gender
plt.figure(figsize=(6,5))
sns.barplot(x='sex', y='total_bill', data=df, palette='Set2' )

plt.title("Average Total Bill by Gender")
```

# Insight 5:
From above graph we can see that average total bill is slightly higher for one gender,
 but overall spending pattern is quite similar between males and females.

```python
# plotting average tip by smoker
plt.figure(figsize=(6,5))
sns.barplot(x='smoker', y='tip', data=df, palette='coolwarm')

plt.title("Average Tip by Smoker")
```

# Insight 6:
From above graph we can see that smokers and non-smokers give nearly similar tips,
 with only slight differences.

# Conclusion:
Most customers are male and non-smokers.  
 Restaurants experience higher footfall during weekends, especially at dinner time.  
The majority of total bills fall in the mid-range category, with tips being smaller but following a similar trend.  
 There is a positive relationship between total bill and tip, and larger groups tend to generate higher bills.  
 Spending patterns are quite similar across genders, and smokers and non-smokers also show comparable tipping behavior.
