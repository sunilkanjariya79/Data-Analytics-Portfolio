#Charts

```python
import matplotlib.pyplot as plt
```

##bar chart:
A Bar Chart uses either horizontal or vertical bars (column chart) to show discrete, numerical comparisons across categories. One axis of the chart shows the specific categories being compared and the other axis represents a discrete value scale.

###horizontal bar chart:
bar_chart (1).png
###vertical bar chart:
bar_chart.png

```python
# vertical bar chart & horizontal bar chart
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(categories, values)
plt.title("Bar Chart")
plt.subplot(1, 2, 2)
plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.tight_layout()
plt.show()
```

##Histogram:
A Histogram visualises the distribution of data over a continuous interval. Each bar in a histogram represents the tabulated frequency at each interval/bin.histogram.png

```python
#histogram
data = [10, 20, 20, 30, 40, 40, 40, 50]
plt.hist(data)
plt.title("Histogram Example")
plt.show()
```

##Column chart:
Column charts are the simplest, most versatile type of visualization used in data analytics. The horizontal chart displays your data in bars proportional to the values they represent.1700739087-column-chart.avif

```python
#column chart
x = ['A','B','C','D']
y = [10,20,15,25]
plt.bar(x, y)
plt.title("Column Chart")
plt.show()
```

##Line chart:
A line chart connects distinct data points through straight lines. Its best use case is to illuminate trends, patterns, and variable changes.1700738749-line-chart.avif

```python
#line chart
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Line Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

##Pie chart:
Extensively used in presentations and offices, Pie Charts help show proportions and percentages between categories, by dividing a circle into proportional segments. Each arc length represents a proportion of each category, while the full circle represents the total sum of all the data, equal to 100%.pie_chart.png

```python
#pie chart
labels = ['Python', 'Java', 'C++', 'JS']
sizes = [30, 25, 20, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart Example")
plt.show()
```

##Donut chart:
A Donut Chart is essentially a Pie Chart but with the area of the centre cut out.

However, Donut Charts have a slight advantage over Pie Charts, which are sometimes criticised for focusing on the relative sizes of the pieces to one another and to the chart as a whole, giving no indication of changes as a whole when compared to other Pie Charts. donut_chart.png

```python
#donut cahrt
labels = ['Python', 'Java', 'C++', 'JS']
sizes = [30, 25, 20, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%',wedgeprops={'width': 0.5})
plt.title("Pie Chart Example")
plt.show()
```

##Stacked bar chart:
A stacked bar chart breaks down and compares parts of a whole across categories. Each bar represents a total value, segmented into colored sections that show how individual components contribute. It’s like a bar chart, but with more layers.1753871026-stacked.avif

```python
#stacked bar chart
x = ['A', 'B', 'C']
y1 = [10, 20, 15]
y2 = [5, 10, 5]
plt.bar(x, y1, label='Part 1')
plt.bar(x, y2, bottom=y1, label='Part 2')
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()
```

##Scatter plot:
Also known as a Scatter Graph, Point Graph, X-Y Plot, Scatter Chart or Scattergram.

A Scatterplot places points on a Cartesian Coordinates system to display all the values between two variables. By having an axis for each variable, you can detect if a relationship or correlation between the two exists.scatterplot.png

```python
#scatter plot
x = [1, 2, 3, 4, 5]
y = [5, 15, 10, 20, 25]
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.show()
```

##Area chart:
An area chart or area graph displays graphically quantitative data. It is based on the line chart. The area between axis and line are commonly emphasized with colors, textures and hatchings. Commonly one compares two or more quantities with an area chart.area_graph.png

```python
#area chart
x=[1,2,3,4]
y=[25,45,45,25]
plt.fill_between(x,y,color='green',alpha=0.5)
plt.title("area chart")
plt.show()
```

##Step chart:
A step chart is a variation of a line chart that uses horizontal and vertical lines to connect data points, emphasizing sudden changes or "steps" in value rather than gradual trends. It’s great for processes or metrics that change at distinct intervals.1749567739-gantt-charts.avif

```python
#step chart
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,25,30]
plt.step(x, y)
plt.title("Step Chart")
plt.show()
```

##Geo chart:
A geo chart maps values across predefined geographic areas—like countries, states, or cities—using colors, bubbles, or markers to represent magnitude.

Use these types of graphs to compare performance across regions, such as sales by country, website traffic by state, or user growth by city.1700739603-geo-chart.avif

```python
#geo chart
plt.style.use('ggplot')
countries = ['India', 'USA', 'UK']
values = [100, 80, 60]
plt.bar(countries, values)
plt.title("Geo (Proxy Chart)")
plt.show()
```

##Heatmap chart:
Heatmap charts are a type of map data visualization that uses a system of color coding to represent value. Each cell in the matrix is assigned a color based on the value it holds.
1700739202-heatmap-charts.avif

```python
#heat map chart
import seaborn as sns
import numpy as np
data = np.random.rand(5,5)
sns.heatmap(data, annot=True)
plt.title("Heatmap")
plt.show()
```

##Box plot(whisker plot):
A Box Plot is also known as a Box and Whisker Plot and it is a graphical tool used to understand the distribution of numerical data. It shows the median, quartiles and possible outliers in a simple visual form. It helps compare distributions between groups with ease.

->multiple groups effectively.

->Identify outliers quickly and visually.

->Reveal data symmetry or skewness at a glance.

->Show variability clearly even for large datasets.

->Provide compact summaries of data using quartiles and range.


Lower Limit=Q1 − 1.5×IQR

Upper Limit=Q3 + 1.5×IQR

->image-3.webpUseful for comparing multiple groups or categories.

```python
#box plot
data = [10, 20, 30, 40, 50]
plt.boxplot(data)
plt.title("Box Plot Example")
plt.show()
```
