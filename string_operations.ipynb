#String operations:


```python
#implementation  of creating and accessing strings
my_string="Hello, Edunet !"
print("Original string:",my_string)
print("First Character:",my_string[0])
print("Last Character:",my_string[-1])
```

```python
#implementation of string concatenation and repetition
str1="Edunet"
str2="Foundation"
concat_str=str1+str2
repeated_str=concat_str*3
print("Concatenated String:",concat_str)
print("Repeated String:", repeated_str)
```

```python
#implementation of string case manipulation
my_string="Hello, Edunet!"
upper_str=my_string.upper()
lower_str=my_string.lower()
title_str=my_string.title()
swapcase_str=my_string.swapcase()
print("Uppercase:",upper_str)
print("Lowercase:",lower_str)
print("Titlecase:",title_str)
print("Swapcase:",swapcase_str)
```

```python
#implementation of searching in strings
substring="Edunet"
found_index=my_string.find(substring)
if found_index!=-1:
  print(f"Substring '{substring}' found at index {found_index}.")
else:
  print(f"Substring '{substring}' not found.")
```

```python
#implementation of replacing substrings
new_string=my_string.replace("Python","World")
print("string after replacement:",new_string)
```

```python
#implementation of string formating
name="Ramar"
age=25
formatted_str=f"My name is {name} and I am {age} year old."
print("Formatted string:",formatted_str)
```

```python
#implementation of Trimming and padding String
str_padding=my_string.center(20,"*")
str_new="            extra spaces            ".strip()
print(str_padding)
print(str_new)
```

```python
#implementation of splitting and joining string
sentence="Python is fun"
words=sentence.split()
joined_sentence='-'.join(words)
print("splitted words:",words)
print("Joined Sentence:",joined_sentence)
```

```python
#implementation of counting characters
char_count=my_string.count("o")
print(f"Character 'o' appears {char_count} times in the string")
```

```python
#implementation of checking string properties
is_alpha="Hello".isalpha()
is_digit="12345".isdigit()
is_alnum="Hello123".isalnum()
is_space="   ".isspace()
print("Is 'Hello' alphabetic ?",is_alpha)
print("Is '12345' a numeric ?",is_digit)
print("Is 'Hello123' alphanumeric ?",is_alnum)
print("Is '   ' all spaces ?",is_space)
```

#List Operations

```python
#implementation of creating a list
fruits = ["apple", "banana", "cherry", "mango", "orange"]
print("Original List:", fruits)
```

```python
#implementation of accessing elements in a list
first_fruit = fruits[0]
last_fruit = fruits[-1]
print("First fruit:", first_fruit)
print("Last fruit:", last_fruit)
```

```python
#implementation of slicing a list
sublist_fruits = fruits[:3]
print("Sublist of first three fruits:", sublist_fruits)
```

```python
#implementation of adding elemetns to a list
fruits.append("grape")
print("List after adding 'grape':", fruits)
```

```python
#implementation of inserting elements into a list
fruits.insert(1, "pineapple")
print("List after inserting 'pineapple' at position 1:", fruits)
```

```python
#implementation of removing elements from a list
fruits.remove("banana")
print("List after removing 'banana':", fruits)
```

```python
#implementation of popping elements from a list
popped_fruit = fruits.pop()
print("Popped fruit:", popped_fruit)
print("List after popping the last element:", fruits)
```

```python
#implementation of finding the lenght of list
length = len(fruits)
print("Number of fruits in the list:", length)
```

```python
#implementation of checking membership in a list
is_in_list = "apple" in fruits
print("Is 'apple' in the list?", is_in_list)
```

```python
#implementation of Iterating over a list
print("Iterating over the list:")
for fruit in fruits:
    print(fruit)
```

```python
#implementation of sorting a list
fruits.sort()
print("List after sorting alphabetically:", fruits)
```

```python
#implementation of reversing a list
fruits.reverse()
print("List after reversing the order:", fruits)
```

```python
#implementation of list comperhensions
long_fruits = [fruit for fruit in fruits if len(fruit) > 5]
print("Fruits with more than 5 letters:", long_fruits)
```

```python
#implementation of copying a list
fruits_copy = fruits.copy()
print("Copied list:", fruits_copy)
```

```python
#implementation of clearing a list
fruits.clear()
print("List after clearing all elements:", fruits)
```

```python
#implementation of extending a list with another list
vegetables = ["carrot", "broccoli", "spinach"]
fruits_copy.extend(vegetables)
print("List after extending with vegetables:", fruits_copy)
```

```python
#implementation of counting occurances of an elements in a list
num_apples = fruits_copy.count("apple")
print("Number of 'apple' in the list:", num_apples)
```

```python
#implementation of  finding the index of an element
if "carrot" in fruits_copy:
    carrot_index = fruits_copy.index("carrot")
    print("Index of 'carrot':", carrot_index)
```

```python
#implementation of removing an element
if len(fruits_copy) > 2:
    del fruits_copy[2]
print("List after removing element at index 2:", fruits_copy)
```

```python
#implementation of nested list
nested_list = [fruits_copy, vegetables]
print("Nested list:", nested_list)
```

#Set Operations


```python
#implementation of creating sets
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
print("Set A:", set_a)
print("Set B:", set_b)
```

```python
#implementation of adding and removing elemetns in a set
set_a.add(6)
set_a.remove(1)
print("Set A after adding 6 and removing 1:", set_a)
print("Set A:", set_a)
print("Set B:", set_b)
```

```python
#implementation of union of sets
union_set = set_a.union(set_b)
print("Union of Set A and Set B:", union_set)
```

```python
#implementation of intersection of sets
intersection_set = set_a.intersection(set_b)
print("Intersection of Set A and Set B:", intersection_set)
```

```python
#implementation of difference of sets
difference_set = set_a.difference(set_b)
print("Difference of Set A and Set B (A - B):", difference_set)
```

```python
#implementation of symmetric difference of sets
sym_diff_set = set_a.symmetric_difference(set_b)
print("Symmetric Difference of Set A and Set B:", sym_diff_set)
```

```python
#implementation of subset and superset testing
set_c = {4, 5}
is_subset = set_c.issubset(set_a)
is_superset = set_a.issuperset(set_c)
print(f"Is Set C a subset of Set A? {is_subset}")
print(f"Is Set A a superset of Set C? {is_superset}")
```

```python
#implementation of set membership testing
is_member = 3 in set_a
print("Is 3 in Set A?", is_member)
```

```python
#implementation of removing duplicate using sets
numbers_list = [1, 2, 3, 4, 4, 5, 6, 6, 7]
unique_numbers = set(numbers_list)
print("List of numbers after removing duplicates:", unique_numbers)
```

```python
#implementation of set comprehension
squares_set = {x ** 2 for x in range(1, 11)}
print("Set of squares of numbers from 1 to 10:", squares_set)
```

```python
#implementation of clearing a set
set_a.clear()
print("Set A after clearing:", set_a)
```

#Dictionary Operations

```python
#implementation of creating a dictionary
student_grades = {
    "Maria": 85,
    "Soumya": 78,
    "Ramar": 62
}
print("Student Grades:", student_grades)
```

```python
#implementation of accessing dictionary values
alice_grade = student_grades["Maria"]
print("Maria's Grade:", alice_grade)
```

```python
#implementation of adding a new key-value pair
student_grades["Rashmi"] = 88
print("Updated Student Grades:", student_grades)
```

```python
#implementation of removing a key-value pair
student_grades.pop("Soumya")
print("Student Grades after removing Soumya:", student_grades)
```

```python
#implementation of checking key membership
is_charlie_in_dict = "Rashmi" in student_grades
print("Is Charlie in the dictionary?", is_charlie_in_dict)
```

```python
#implementation of iterating over a dictionary(key and values)
print("Iterating over student names and grades:")
for student, grade in student_grades.items():
    print(f"{student}: {grade}")
```

```python
#implementation of using the get() method
david_grade = student_grades.get("Ramar", "Not found")
print("Ramar's Grade:", david_grade)
```

```python
#implementation of merging dictionaries
additional_grades = {"Bhawani": 60, "Usha": 75}
student_grades.update(additional_grades)
print("Student Grades after merging with additional data:", student_grades)
```

```python
#implementation of dictionary comprehension
squared_numbers = {x: x ** 2 for x in range(1, 6)}
print("Dictionary of numbers and their squares:", squared_numbers)
```

```python
#implementation of handling nested dictionaries
nested_dict = {
    "USA": {"New York": 8000000, "Los Angeles": 4000000},
    "India": {"Mumbai": 20000000, "Delhi": 15000000}
}
print("Nested dictionary of countries and cities:", nested_dict)

ny_population = nested_dict["USA"]["New York"]
print("Population of New York:", ny_population)
```

```python
#implementation of clearing and copying dictionaries
student_grades_copy = student_grades.copy()
print("Copy of student grades:", student_grades_copy)

student_grades.clear()
print("Student grades after clearing:", student_grades)
```
