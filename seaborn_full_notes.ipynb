## What is Seaborn
Seaborn is a Python data visualization library built on top of Matplotlib. It is used to create attractive and statistical graphics easily.

## Basic Setup

```
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

## Bar Chart
### sns.barplot()
Used for categorical comparison.

```
sns.barplot(x=['A','B','C'], y=[10,20,15], palette='viridis')
plt.title('Bar Chart')
plt.show()
```

## Line Chart
### sns.lineplot()

```
sns.lineplot(x=[1,2,3], y=[10,20,15], marker='o')
plt.title('Line Chart')
plt.show()
```

## Scatter Plot
### sns.scatterplot()

```
sns.scatterplot(x=[1,2,3], y=[10,20,15])
plt.title('Scatter Plot')
plt.show()
```

## Histogram
### sns.histplot()

```
sns.histplot([10,20,20,30], kde=True)
plt.title('Histogram')
plt.show()
```

## Box Plot
### sns.boxplot()

```
sns.boxplot(data=[10,20,30,40])
plt.title('Box Plot')
plt.show()
```

## Violin Plot
### sns.violinplot()

```
sns.violinplot(data=[10,20,30,40])
plt.title('Violin Plot')
plt.show()
```

## Heatmap
### sns.heatmap()

```
import numpy as np
data = np.random.randint(1,100,(5,5))
sns.heatmap(data, annot=True)
plt.title('Heatmap')
plt.show()
```

## Common Properties
### Used in all charts
- title
- xlabel
- ylabel
- palette
- hue

## Conclusion
Seaborn simplifies data visualization and provides better graphical representation than matplotlib.
